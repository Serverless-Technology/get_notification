```python
import json
def save_to_json(event):
    file_name=event['title']
    with open(f'{file_name}.json', 'w', encoding='utf-8') as json_file:
        json.dump(event, json_file, ensure_ascii=False, indent=4)

for event in events_results:
    print(event)
    save_to_json(event)
```
